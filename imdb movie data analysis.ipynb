```python
#Classify Movies Based on Ratings [Excellent, Good, and Average]
```

```python
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "excellent"
    elif rating>=6.0:
        return "good"
    else:
        return "avg"
```

```python
data['rating_cat']=data['Rating'].apply(rating)
```
